```python
# importando biblioteca Pyautogui (controla mouse e teclado)
import pyautogui as py

#importando biblioteca time (aguardar tempo): time.sleep(1)
import time
```

# Alguns dos principais funções do python.

pyautogui.write() -> Escrever.

pyautogui.click() -> Clicar com mouse.

pyautogui.locateOnScreen -> identifica uma imagem na sua tela.

pyautogui.hotkey -> usa atalhos do teclado.

pyautogui.press -> aperta um botão do teclado.

# Inicio da automação.

```python
# importando biblioteca Pyautogui
import pyautogui as py


# 1 abrindo o gmail:
py.press("win")


# 2 digitar chrome
py.write("chrome")

#dá um segundo para executar a proxima função.
time.sleep(1)

# pressionar tecla enter
py.press("Enter")






#aguardar 3 segundo:
time.sleep(1)

#pesquisando gmail:
py.write("https://mail.google.com/mail/u/0/#inbox")

#prssionando enter:
py.press("enter")

#aguardar 1 segundo:
time.sleep(3)






# para clicar em um campo especifico. # use a ferramenta lightshot para tirar print:
# Agora tire o prit somente do local onde vamos clicar e salve o arquivo dentro da mesma pasta onde está nosso código.
# Salve o arquivo sempre em PNG.





# localizando a imagem e clicar no meio da imagem. #pyautogui.locateonscreen("nom da imagem")
x, y, largura, altura = py.locateOnScreen("escrever gmail.png")
# quando formos localizar a imagem  python vai nos dar 4 informações:
#Pisição , posição y, largura e altura.

# Clicar no meio da imagem:
py.click(x + largura/2, y + altura/2)






#escreve menssagem para mim:
py.write("forcecell954@gmail.com")

#pressionar enter
py.press("enter")

# reconhecimento campo assunto:
x, y, largura, altura = py.locateOnScreen("assunto gmail.png")

#clicar no local:
py.click(x + largura/2, y + altura/2)


#toda vez que formos digitar uma menssagem devemos utilizar aspas X3
texto =f''' 
ola tenha um excelente dia de trabalho
'''

#escrevendo menssagem
py.write(texto)

#clicando no botão de enviar.
x, y, largura, altura = py.locateOnScreen("enviar.png")

#clicar na imagem:
py.click(x + largura/2, y + altura/2)




#esperar gmail
#enquanto a logotipo não existir na tela retorne none.
#while not py.locateOnScreen("escrever gmail.png"):
 #   time.sleep(1)


```
